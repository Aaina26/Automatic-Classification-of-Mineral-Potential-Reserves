```python
from keras.preprocessing.image import ImageDataGenerator
from skimage import io
```

```python
datagen=ImageDataGenerator(
    rotation_range=45,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='reflect')
```

```python
import numpy as np
import os
from PIL import Image
```

```python
image_directory='/content/drive/MyDrive/Data2(Augmented)/Manganese'
SIZE=256
dataset=[]
```

```python
import cv2
from keras.utils.image_utils import img_to_array
from keras.utils.image_dataset import load_image
my_images=os.listdir(image_directory)
for i, image_name in enumerate(my_images):
  img_array = cv2.imread(os.path.join(image_directory,image_name) ,cv2.IMREAD_GRAYSCALE)  # convert to array
  new_array = cv2.resize(img_array, (SIZE,SIZE))
  dataset.append(np.array(new_array))
```

```python
dataset = np.array(dataset).reshape(-1, SIZE, SIZE, 1)
dataset.shape
```

```python
i=0
for batch in datagen.flow(dataset,batch_size=16,
                          save_to_dir='/content/drive/MyDrive/Data2(Augmented)/Manganese',
                          save_prefix='aug',
                          save_format='tiff'):
  i+=1
  if i>50:
    break
```
